```python
days = df.resample('3D',on = 'Date & Time').mean().reset_index()
print(days.info())
```

```python
Comp = days[['AC 1','AC 2','AC 3']] \
    .plot(style='-', figsize=(15, 5), title='Comparison between  AC 1, AC 2, AC 3')
```

### Checking for collinearity among attributes of a dataset Using Scatter_Matrix

scatter_matrix() can be used to easily generate a group of scatter plots between all pairs of numerical features. It creates a plot for each numerical feature against every other numerical feature and also a histogram for each of them.

```python
from pandas.plotting import scatter_matrix
attributes = ['AC 1', 'AC 2', 'AC 3', 'AC 4']
scatter_matrix(df[attributes], figsize=(12,8),color='b')
```
